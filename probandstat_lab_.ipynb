```python
#Program 1
import numpy as np
val=list(map(float,input().split()))
def mean_value(data):
    sum=0
    for i in data:
        sum=sum+i
    return(sum/len(data))
mean=mean_value(val)
print("the mean of given data is ",mean)
def var_value(data,mean):
    ss=0
    for i in data:
        ss=ss+(i-mean)**2
    return(ss/len(data))
var=var_value(val,mean)
print("the varience of given data is ",format(var,'.6'))
```

```python
#program 2:
import numpy as np
x=list(map(float,input().split()))
f=list(map(float,input().split()))
def mean(x,f):
    s=0
    for i,j in zip(x,f):
        s=s+i*j
    return(s/sum(f))
m=mean(x,f)
print("the mean is ",m)
def var(x,f,m):
    ss=0
    for i,j in zip(x,f):
        ss=ss+(i-m)**2*j
    return(ss/sum(f))
v=var(x,f,m)
#print("the varience is ",v)
print("the standaed devuation is",format(np.sqrt(v),'.5'))
```

```python
#program 4:
import numpy as np
l=list(map(float,input().split()))
u=list(map(float,input().split()))
f=list(map(float,input().split()))
n=sum(f)

cumsum=0
for i in range(len(f)):
    cumsum=cumsum+f[i]
    if cumsum>n/2:
        low=l[i]
        cf=cumsum-f[i]
        fre=f[i]
        h=u[i]-l[i]
        median=low+(n/2-cf)*h/fre
        break
print("median is",median)
i=f.index(max(f))
f1=f[i]
f0=f[i-1]
f2=f[i+1]
h=u[i]-l[i]
mode=low+(f1-f0)*h/(2*f1-f0-f2)
print("mode is",format(mode,'.7'))
```

```python
#program 1:
from math import comb

def binomial_probability(N, k, p):
    return comb(N, k) * (p ** k) * ((1 - p) **(N - k))

def average_users(N, p):
    return N * p

N = int(input("Enter the total number of users (N > 70): "))
p = 0.25

k1 = 40
prob_40 = binomial_probability(N, k1, p)
print(f"Probability that exactly 40 users will not skip the ad: {prob_40:.6f}")

prob_at_least_60 = sum(binomial_probability(N, k, p) for k in range(60, N + 1))
print(f"Probability that at least 60 users will not skip the ad: {prob_at_least_60:.6f}")

average_watchers = average_users(N, p)
print(f"Average number of users who will watch the ad completely: {average_watchers:.2f}")
```

```python
#program 1:
import math
mu = 2
def poisson_pmf(mu,x):
    return ((math.e)**(-mu))*(mu**x)/(math.factorial(x))
def poisson_cdf(mu,x):
    cdf=0
    for i in range(0,x):
        cdf+=poisson_pmf(mu,i)
    return cdf
print("(i)",format(poisson_pmf(mu,10),'.8'))
print("(ii)",1-poisson_cdf(mu,15))
print("(iii)",1-poisson_cdf(mu,21))
```

```python
#program 2:
from scipy.stats import norm
mu=50000
sig=20000
def normal_cdf(mu,sig,x):
    z=(x-mu)/sig
    prob=norm.cdf(z)
    return prob
print("(i)",format(normal_cdf(mu,sig,40000)*100,'.5'))
print("(ii)",format((normal_cdf(mu,sig,65000)-normal_cdf(mu,sig,45000))*100,'.5'))
print("(iii)",format((1-normal_cdf(mu,sig,70000))*100,'.5'))
```
